```r
library(tidyverse)
```

```r
wfm_all = read_tsv('data/wholefoods/wfm_final.tsv', col_types="cccccccDddddcccc")
head(wfm_all,1)
#summary(wfm_all)
```

```r
wfm_stores_by_state = wfm_all %>%     
    group_by(StateAbr) %>%
    distinct(CustNum) %>%
    select(-CustNum) %>%
    summarize(count=n()) %>%
    arrange(-count) 
wfm_stores_by_state
```
